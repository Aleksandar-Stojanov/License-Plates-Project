```python
import pytesseract
import matplotlib.pyplot as plt
import cv2
import glob
import os
import easyocr
import numpy as np
import imutils
from difflib import SequenceMatcher
```

```python
def calculate_predicted_similarity(actual_list, predicted_list):
    similarities = []

    for actual_plate, predict_plate in zip(actual_list, predicted_list):
        similarity_ratio = SequenceMatcher(None, actual_plate, predict_plate).ratio()
        similarities.append(similarity_ratio)

    return similarities
```

```python
def calculate_cumulative_similarity(actual_list, predicted_list):
    total_similarity = 0

    for actual_plate, predict_plate in zip(actual_list, predicted_list):
        similarity_ratio = SequenceMatcher(None, actual_plate, predict_plate).ratio()
        total_similarity += similarity_ratio

    average_similarity = total_similarity / len(actual_list)

    return average_similarity
```

```python
import csv



csv_file_path = '/content/drive/MyDrive/Colab Notebooks/LicensePlatesProjectNew2/lpr.csv'
labels_dict = {}

with open(csv_file_path, 'r') as csvfile:
    csvreader = csv.reader(csvfile)
    next(csvreader)
    for row in csvreader:
        image_name = row[1]
        label = row[2]
        labels_dict[image_name] = label

input_folder = '/content/drive/MyDrive/Colab Notebooks/LicensePlatesProjectNew2/cropped_lps/cropped_lps'
path_for_license_plates = os.path.join(input_folder, '*.jpg')

num_images_to_process = 3000
processed_images_count = 0

best_psm_value = 0
best_accuracy = 0
best_prediction = []

for i in range(6, 9):
    processed_images_count = 0
    list_license_plates = []
    predicted_license_plates = []
    for path_to_license_plate in glob.glob(path_for_license_plates):

        input_path = path_to_license_plate

        img = cv2.imread(input_path)

        if img is not None:
          gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
          img1 = cv2.bilateralFilter(gray, 11, 17, 17)
          text = pytesseract.image_to_string(img1, lang='eng', config='--oem 3 --psm 6 -c tessedit_char_whitelist=ABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789')

          if not (text):
            continue

          predicted_result = pytesseract.image_to_string(
              img1, lang='eng', config=f'--oem 3 --psm {i} -c tessedit_char_whitelist=ABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789'
          )

          filter_predicted_result = "".join(predicted_result.split()).replace(":", "").replace("-", "")
          predicted_license_plates.append(filter_predicted_result)

          image_name = os.path.basename(input_path)
          true_label = labels_dict.get(image_name, '')

          list_license_plates.append(true_label)

          processed_images_count += 1
          print(processed_images_count)
          if processed_images_count >= num_images_to_process:
              break


    average_similarity = calculate_cumulative_similarity(list_license_plates, predicted_license_plates)

    if best_accuracy < average_similarity:
        best_psm_value = i
        best_accuracy = average_similarity
        best_prediction = [i for i in predicted_license_plates]

print(f"Best PSM value: {best_psm_value}")
print(f"Best Accuracy: {best_accuracy}")

```

```python
similarities = calculate_predicted_similarity(list_license_plates, best_prediction)
for actual_plate, predict_plate, similarity in zip(list_license_plates, best_prediction, similarities):
    print(f"Actual: {actual_plate}\tPredicted: {predict_plate}\tSimilarity: {similarity:.2f}")

average_similarity = calculate_cumulative_similarity(list_license_plates, best_prediction)

print(f"Average Similarity: {average_similarity:.2%}")
```
